**import the necessary libraries.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf

# 1. Charger le fichier
filename = 'morocco_weather.csv'

# On cherche la ligne où commencent les données
start_line = 0
with open(filename, 'r') as f:
    for i, line in enumerate(f):
        if 'temperature_2m_max' in line:
            start_line = i
            break

df = pd.read_csv(filename, skiprows=start_line)

# 2. Nettoyer les noms des colonnes (enlever les unités comme °C)
df.columns = [c.split(' ')[0] for c in df.columns]

# 3. Convertir la date
df['time'] = pd.to_datetime(df['time'])

print(f"✅ {len(df)} jours de météo chargés !")
print(df[['time', 'temperature_2m_max']].head())
```

**Étape 2 : Visualisation (Vérifie tes données)**

```python
plt.figure(figsize=(15, 5))
plt.plot(df['time'], df['temperature_2m_max'], color='red', linewidth=0.5)
plt.title('Historique Température Max au Maroc (2018-2026)')
plt.ylabel('Degrés Celsius')
plt.show()
```

**Étape 3 : Préparer "l'intelligence" de l'IA**

```python
# Sélection des variables
features = ['temperature_2m_max', 'temperature_2m_min', 'precipitation_sum']
data = df[features].values

# Mise à l'échelle (0 à 1) pour aider l'IA
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

def create_sequences(data, lookback=30, forecast=30):
    X, y = [], []
    for i in range(len(data) - lookback - forecast):
        X.append(data[i : i + lookback])
        y.append(data[i + lookback : i + lookback + forecast, 0])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data)
print(f"Structure des données : {X.shape}")
```

**Étape 4 : Entraînement du modèle (LSTM)**
C'est ici que la magie opère. L'IA va faire 20 passages sur tes données pour apprendre les cycles.

```python
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(64, return_sequences=True, input_shape=(30, 3)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(30) # Prédiction de 30 jours
])

model.compile(optimizer='adam', loss='mse')

print("🚀 Entraînement en cours...")
model.fit(X, y, epochs=20, batch_size=32, verbose=1)
print("✅ IA entraînée avec succès !")
```

```python
# 1. On récupère les 30 derniers jours de ton fichier CSV pour lancer la prédiction
last_30_days = scaled_data[-30:]
last_30_days_input = np.expand_dims(last_30_days, axis=0) # On ajuste le format pour l'IA

# 2. L'IA prédit les 30 prochains jours
prediction_scaled = model.predict(last_30_days_input)

# 3. On "dé-normalise" le résultat pour repasser de (0-1) à des degrés Celsius
# Note: On crée un tableau vide pour aider le scaler à faire la conversion inverse
dummy_scaler = np.zeros((30, 3))
dummy_scaler[:, 0] = prediction_scaled[0]
prediction_final = scaler.inverse_transform(dummy_scaler)[:, 0]

# 4. Affichage des résultats
print("--- PRÉVISIONS POUR LES 30 PROCHAINS JOURS AU MAROC ---")
for i, temp in enumerate(prediction_final[:7]): # On affiche la 1ère semaine
    print(f"Jour {i+1} : {temp:.2f} °C")

# 5. Graphique de comparaison
plt.figure(figsize=(12, 6))
plt.plot(range(-30, 0), df['temperature_2m_max'].values[-30:], label="Passé (30 derniers jours)", color='blue')
plt.plot(range(0, 30), prediction_final, label="Prédiction IA (30 prochains jours)", color='red', linestyle='--')
plt.axvline(x=0, color='black', linestyle='-') # Ligne de séparation aujourd'hui
plt.title("Test de l'IA : Transition entre le passé et le futur (Casablanca)")
plt.ylabel("Température Max (°C)")
plt.xlabel("Jours (0 = Aujourd'hui)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf

# 1. Préparation du convertisseur
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# 2. Options magiques pour corriger l'erreur LSTM
converter.target_spec.supported_ops = [
    tf.lite.OpsSet.TFLITE_BUILTINS, # Opérations TFLite standards
    tf.lite.OpsSet.SELECT_TF_OPS    # Autoriser les opérations TF nécessaires pour le LSTM
]

# Désactiver l'option qui pose problème avec les LSTM
converter._experimental_lower_tensor_list_ops = False

# 3. Conversion
try:
    tflite_model = converter.convert()

    # 4. Sauvegarde
    model_name = 'maroc_weather_model.tflite'
    with open(model_name, 'wb') as f:
        f.write(tflite_model)

    print(f"✅ Succès ! Le fichier '{model_name}' a été généré avec les options compatibles LSTM.")

    # 5. Téléchargement
    from google.colab import files
    files.download(model_name)

except Exception as e:
    print(f"❌ Erreur persistante : {e}")
```
